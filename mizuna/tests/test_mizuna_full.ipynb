```python
import pandas as pd
import seaborn as sns
from mizuna import Mizuna
```

```python
df = pd.read_csv('data/gold_data.csv')
df.head()
```

```python
df[['Year', 'Month']] = df['Date'].str.split('-', expand=True)
df['Year'] = df['Year'].astype('int64')
df['Month'] = df['Month'].astype('int64')
df.info()
```

```python
# plot = sns.scatterplot(x='Year', y='Price', hue='Month', data=df)
plot = sns.scatterplot(x='Year', y='Price', data=df)
plot.figure.savefig('figures/price_over_year.pdf', format='pdf')
```

```python
plot = sns.scatterplot(x='Year', y='Price', hue='Month', data=df)
# plot = sns.scatterplot(x='Year', y='Price', data=df)
plot.figure.savefig('figures/price_over_year2.pdf', format='pdf')
```
